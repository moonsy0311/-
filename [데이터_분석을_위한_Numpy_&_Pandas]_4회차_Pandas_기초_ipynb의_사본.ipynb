## 2.1. Series란 무엇인가?

Series는 Pandas의 기본 데이터 구조 중 하나로, 1차원 배열과 비슷한 형태를 가집니다. 각 데이터는 인덱스를 가지며, 인덱스를 통해 데이터를 참조할 수 있습니다. Series는 동일한 데이터 타입의 값들을 담을 수 있으며, 숫자, 문자열, 부울 등의 데이터를 포함할 수 있습니다.

**주요 특징**
* 일차원 배열: 단일 데이터 타입의 데이터를 순차적으로 저장
* 인덱스: 각 데이터에 고유한 인덱스를 부여하여 데이터 접근 가능
* 다양한 데이터 타입: 숫자, 문자열, 날짜 등 다양한 데이터 타입 저장 가능

## 2.2. Series 생성 방법


Series는 여러 가지 방법으로 생성할 수 있습니다. 리스트, 딕셔너리, numpy 배열 등을 사용하여 Series를 만들 수 있습니다.



### 2.2.1. 리스트로부터 생성



```python
data = [10, 20, 30, 40, 50]
series = pd.Series(data) #대문자 주의하기
series
```

### 2.2.2. 딕셔너리로부터 생성

```python
data = {'a': 10, 'b': 20, 'c': 30, 'd': 40, 'e': 50} #dic으로 정의해놓으면 인덱스가 key값으로 나옴 (숫자가 아닌)
series = pd.Series(data)
series
#series[0] = 10
```

#### 2.2.3. 인덱스지정

```python
data = [10, 20, 30, 40, 50]
series = pd.Series(data, index=['a', 'b', 'c', 'd', 'e'])  #인덱스 설정을 다시 해놓으면 인덱스가 0부터 시작하는 숫자가 아닌 사용자 지정으로 변경 가능 / 근데 거의 안씀
series
```

## 2.3. Series의 인덱싱과 슬라이싱


### 인덱싱

```python
series[0]    # 첫 번째 요소
```

```python
series['a']  # 인덱스 'a'의 값
```

### 슬라이싱

```python
series[1:3]  # 두 번째 요소부터 세 번째 요소까지
```

```python
series['a':'c']  # 인덱스가 'a'부터 'c'까지의 요소
```

## 2.4. DataFrame이란 무엇인가?



Pandas의 DataFrame은 행과 열로 구성된 "이차원 데이터" 구조로, 테이블 형식의

1.   항목 추가
2.   항목 추가

데이터를 다루기에 적합합니다. 각 열은 Series 객체로 구성되며, 다양한 데이터 타입을 포함할 수 있습니다.

**주요 특징**
* 이차원 데이터 구조: 행과 열로 구성된 테이블 형태
* 다양한 데이터 타입: 숫자, 문자열, 날짜 등 다양한 데이터 타입 포함 가능
* 유연한 데이터 조작: 데이터 선택, 필터링, 정렬, 그룹화 등의 다양한 데이터 조작 기능 제공

## 2.4. DataFrame 생성 방법

### 2.4.1. 리스트로부터 생성

```python
data = [
    ['Alice', 25],
    ['Bob', 30],
    ['Charlie', 35]
]
```

```python
data
```

```python

df = pd.DataFrame(data, columns=['Name', 'Age'])
df
```

### 2.4.2. 딕셔너리로부터 생성



```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35]
}
df = pd.DataFrame(data)
df
```

series dic => 단일 값 ('a';10, 'b':20)

dataframe dic => 여러 값 ('name';'alice, bob, charlie')

-> dataframe은 dic의 value값의 개수가 같아야 설정가능

```python
df['Name']  # 'Name' 열 선택
```

```python
df.iloc[0]  # 첫 번째 행 선택 (idx값이 0인 행)
```

```python
df.loc[0]  # 인덱스 숫자가 0인 행 선택
```
