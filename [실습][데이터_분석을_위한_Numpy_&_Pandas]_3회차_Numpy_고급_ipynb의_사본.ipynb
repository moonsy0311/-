````python
# 정규 분포에서 1만개 데이터를 생성
normal_array = np.random.normal(0, 1, size=(100, 100))

# 시각화 라이브러리
import matplotlib.pyplot as plt
# 히스토 그램을 그림
plt.hist(normal_array.flatten(), bins=50, edgecolor='black') # 히스토 그램 그림
plt.show() # 최종적으로 그림을 출력해라
````

`plt.hist(normal_array.flatten(), bins=100, edgecolor='black')`는 정규 분포에서 생성된 난수로 이루어진 1차원 배열의 히스토그램을 그리는 코드입니다.

* bins=100:  
    히스토그램의 막대(bin)의 수를 지정하는 인자입니다. bins=100은 히스토그램을 100개의 막대로 나누겠다는 의미입니다. 막대의 수가 많을수록 데이터의 세부적인 분포를 볼 수 있지만, 너무 많으면 해석이 어려울 수 있습니다.

* edgecolor='black':  
    히스토그램 막대의 테두리 색상을 지정하는 인자입니다. 여기서는 막대의 테두리 색상을 검정색(black)으로 설정하여 막대를 더 잘 구분할 수 있도록 합니다.

## 파일 입출력

### 배열 저장하기

Numpy 배열을 .npy 파일로 저장합니다. 이 파일은 Numpy 배열 데이터를 효율적으로 저장할 수 있도록 설계된 바이너리 파일 형식입니다.



````python
# 배열 생성
a = np.array([[1, 2, 3], [4, 5, 6]])
a
````

````python
a.shape
````

````python
# 배열 저장
np.save('array.npy', a)
````

여러 배열을 하나의 파일에 저장합니다. .npz 파일 형식으로 저장됩니다.



````python
# 배열 생성
a = np.array([[1, 2, 3], [4, 5, 6]])
b = np.array([7, 8, 9])

# 여러 배열 저장
np.savez('arrays.npz', array1=a, array2=b)
````

텍스트 파일로 배열을 저장합니다. CSV 파일 형식으로 저장할 수 있습니다.



````python
# 배열 생성
a = np.array([[1, 2, 3], [4, 5, 6]])

# 배열 텍스트 파일로 저장
np.savetxt('array.txt', a, delimiter=',')
````

````python
np.savetxt('array.csv', a, delimiter=',')
````

## 뷰와 복사

Numpy에서는 배열의 데이터를 공유하는 뷰(view)와 데이터를 복사하는 복사(copy) 기능을 제공합니다. 이를 통해 메모리 효율성과 데이터 무결성을 적절히 관리할 수 있습니다.



### 뷰(View)

````python
# 원본 배열 생성
a = np.array([1, 2, 3, 4, 5])
a
````

````python
# 배열 뷰 생성
b = a[1:4]
b
````

````python
# 뷰 수정
b[0] = 10
b
````

````python
# 원본 배열 확인
a
````

reshape과 ravel은 배열의 뷰를 생성하는 대표적인 함수입니다.

````python
# 원본 배열 생성
a = np.array([[1, 2, 3], [4, 5, 6]])
a
````

````python
reshaped = a.reshape((3, 2))
reshaped
````

````python
reshaped[0, 0] = 10
reshaped
````

````python
a
````

### 배열 복사

복사는 원본 배열의 데이터를 복제하여 새로운 배열을 생성합니다. 복사된 배열을 수정해도 원본 배열에는 영향을 미치지 않습니다.



````python
# 원본 배열 생성
a = np.array([1, 2, 3, 4, 5])
a
````

````python
# 배열 복사
b = a[1:4].copy()
b
````

````python
# 복사된 배열 수정
b[0] = 10
b
````

````python
a
````
